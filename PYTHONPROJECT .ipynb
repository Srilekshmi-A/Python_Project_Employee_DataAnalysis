## 1) How many are there in each Team and the percentage splitting with respect to the total employees?

```python
count=df['Team'].value_counts()
percentage=round((count/sum(count)*100),2)
df1=pd.DataFrame({'Team':count.index,'Count':count.values,'percentage':percentage.values})
df1
```

##### The team with more number of employees is Team 'New Orleans Pelicans' with 19 employees making 4.15% of the total employees.
##### 'Orlando Magic'	and 'Minnesota Timberwolves' are the teams with least number of employees .Both team consists of 14 employees, each makes 3.06% .

##### This dataframe shows the details of employees grouped according to their position

```python
a=df['Position'].value_counts()
plt.bar(a.index,a,color='lightseagreen',width=0.5)
plt.xlabel('Position')
plt.ylabel('Count of employees')
plt.title("Count of employees in each position")
plt.show()
```

# 5) Find if there is any correlation between age and salary , represent it visually

```python
sns.scatterplot(x="Age", y="Salary", hue="Age", data=df)

plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("age v/s salary")

plt.show()
```

##### From this scatterplot we cannot say that there is a positive correlation between salary and age,as we can see in age group 35-40 salary is less mor majority of people.At the same time highest salary also belongs to age grop 35-40
##### In all the age groups,there are employees with minimum salary and they form majority of the employees.
##### So we can say that not only Age can determine the salary ,there can be many other factors also.
